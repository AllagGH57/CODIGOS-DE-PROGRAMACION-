```python
suma=0
num=int(input("cuantas calificaciones vas a introducir?" ))
for i in range(num):
  calificacion=float(input("dame la calificacion: "))
  suma=suma+calificacion
  promedio=suma/num
  print("este es el promedio de tus calificaciones:",promedio)
```
